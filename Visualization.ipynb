```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from pathlib import Path

path_Current_Directory = Path.cwd()
path_Data_Directory = path_Current_Directory / "Data"
path_Forecast_DF = path_Data_Directory / "Forecast.xlsx"

df_forecast = pd.DataFrame()
df_forecast = pd.read_excel((path_Forecast_DF), keep_default_na=False)

df_filter = df_forecast.loc[(df_forecast["iso_code"].isin(["POL","UKR","ZAF","UKR-frcst","POL-frcst","RUS", "RUS-frcst"]) & (df_forecast["date"] >= "2020-04-01"))]
df_pvt_smooth = df_filter.pivot(index="date", columns="iso_code", values="new_cases_smoothed")
df_pvt_exact = df_filter.pivot(index="date", columns="iso_code", values="new_cases")

df_pvt_smooth.plot(subplots=False, figsize=(24,10), title="Зареєстрована кількість нових випадків захворювання CoVID-19 (середня кількість за 7 днів)", grid=True, style=["-","--","-","-.","-","-.",":"], xlabel="дата", ylabel="середня кількість випадків за 7 днів", kind="line")
```

```python
df_pvt_exact.plot(subplots=False, figsize=(24,10), title="Зареєстрована кількість нових випадків захворювання CoVID-19", grid=True, style=["-","--","-","-.","-","-.",":"], xlabel="дата", ylabel="кількість випадків", kind="line")
```

```python
df_pvt_exact.plot(subplots=False, figsize=(24,10), title="Зареєстрована кількість нових випадків захворювання CoVID-19", grid=True, sharey=True, xlabel="дата", ylabel="кількість випадків", kind="box")
```
